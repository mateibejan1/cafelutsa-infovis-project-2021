```python
#read the dataset
import pandas as pd

data = pd.read_csv(data_dir + "Indicators.csv", delimiter=',')
print(data.keys())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Choose Arab World, Caribbean small states, North America, Euro area
df = pd.DataFrame([data.loc[13585], data.loc[13586], data.loc[13590], data.loc[13608]])
df = df.reset_index()
plt.figure()
plt.ylim(500, 50000)
sns.regplot(data = df, x = df["Prevalence of HIV, total (% of population ages 15-49)"], y = df["GDP per capita (current US$)"])
for i in range(df.shape[0]):
  plt.text(x = df["Prevalence of HIV, total (% of population ages 15-49)"][i], y = df["GDP per capita (current US$)"][i], s = df["CountryName"][i])

plt.show()
```
